```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


dt=pd.read_csv(r"C:\Users\msi\Documents\supcom\MLS\studentPerformance.csv")
x=dt.values[:,:-2]
y=dt.values[:,3]
z=dt.values[:,4]
x_train,x_test,y_train,y_test,z_train,z_test=train_test_split(x,y,z,test_size=0.3,random_state=2)
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
training_res=lr_model.score(x_train,y_train)
test_res=lr_model.score(x_test,y_test)
print(f"R^2:{training_res}")
print(f"R^2:{test_res}")
reg_log = LogisticRegression(solver='liblinear', random_state=42)
reg_log.fit(x_train, z_train)
from sklearn.metrics import accuracy_score, classification_report
z_log_pred = reg_log.predict(x_test)

# Prédiction des PROBABILITÉS (très utile en logistique)
z_log_proba = reg_log.predict_proba(x_test)[:, 1] # Probabilité de la classe 1

# Évaluation (exemple avec la précision et le rapport de classification)
accuracy = accuracy_score(z_test, z_log_pred)

print(f"\n**LogisticRegression Résultats**")
print(f"Précision (Accuracy): {accuracy:.3f}")
print(f"Rapport de classification:\n{classification_report(z_test, z_log_pred, target_names=["Fail (0)", "Pass (1)"])}")
# ⚡ Prédiction pour un utilisateur
# ============================
print("\n=== Prédiction Étudiant ===")
study_hours = float(input("Entrez Study Hours (0-10): "))
attendance = float(input("Entrez Attendance (50-100%): "))
practice_tests = int(input("Entrez Practice Tests (0+): "))

# Créer un DataFrame avec les entrées
user_data = pd.DataFrame([[study_hours, attendance, practice_tests]],
                         columns=["Study_Hours", "Attendance", "Practice_Tests"])

# Prédiction score final (régression linéaire)
pred_score = lr_model.predict(user_data)[0]

# Prédiction Pass/Fail (régression logistique)
pred_pass = reg_log.predict(user_data)[0]
pred_proba = reg_log.predict_proba(user_data)[0][1]  # proba d'être en classe 1 (Pass)

# ============================
# Affichage du résultat
# ============================
print(f"\nPrédiction du Score Final : {pred_score:.2f}/100")
print(f"Probabilité de réussir : {pred_proba*100:.1f}%")
if pred_pass == 1:
    print("✅ Résultat : PASS (réussi)")
else:
    print("❌ Résultat : FAIL (échoué)")
```
